```python
!pip install yahoofinancials
from yahoofinancials import YahooFinancials
import pandas as pd
```

```python
def get_bspl(ticker):
  yf = YahooFinancials(ticker)
  data1 = yf.get_financial_stmts("quarterly", "balance")["balanceSheetHistoryQuarterly"][ticker][0]
  df_bs = pd.DataFrame.from_dict(data1 )
  data2 = yf.get_financial_stmts("quarterly", "income")["incomeStatementHistoryQuarterly"][ticker][0]
  df_pl = pd.DataFrame.from_dict(data2)
  
  df_bspl = pd.concat([df_bs, df_pl]) 
  df_bspl = df_bspl.rename(columns={df_bspl.columns[0]:ticker})

  return df_bspl.drop_duplicates()
```

```python
_tickers1 = ["V","PYPL","VZ","T","VIV","AMX","VOD","TU","TEF","TSLA","GM","F","STLA","HMC","NIO","PG","KO","PEP","ABEV","NKE","PM","UL","BUD","MDLZ","MO"]
_tickers2 = ["DIS","CMCSA","SBUX","WBD","BSX","HLN","AAPL","NVDA","TSM","CSCO","QCOM","RTX","AMD","INTC","MU","NOK","XOM","PBR","CVX","SHEL","COP","TTE","BP","EQNR","CNQ"]
_tickers3 = ["SU","UBS","BRK.B","JPM","BAC","ITUB","WFC","MS","HSBC","SCHW","BBD","JNJ","MRK","ABBV","PFE","NVO","NVS","AZN","ABT","BMY","SNY","ENB","SLB","EPD","KMI"]
_tickers4 = ["ET","VALE","BHP","RIO","FCX","GOLD","SUZ","AMCR","GE","ABB","AMZN","WMT","HD","BABA","PDD","CVS","JD","TJX","CPNG","MSFT","GOOG","GOOGL","META","ORCL","BEKE"]
_tickers5 = ["SAP","INFY","WIT","NU","CSX","UBER","GRAB","NEE","SO","PCG"]

tickers = _tickers1 ### <---

df = pd.DataFrame()
i_count=0
i_total = len(tickers)
for e in tickers:
  i_count += 1
  if i_count == 1:
      print("Progress: ")
  print(">>> {}/{}: {}".format(str(i_count), str(i_total), e), end="; ")
  try:
    df = df.join(get_bspl(e), how="outer", sort=False)
  except Exception as e:
    print(">>> > Error: " + str(e))
```
