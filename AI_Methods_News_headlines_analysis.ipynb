```python
from textblob import TextBlob
import numpy as np
from pyfinviz.quote import Quote
import pandas as pd
```

```python
portfolio = ["AAPL", "KO", "MCD", "PFE", "MSFT", "TSM", "TSLA", "MRK", "PM", "JNJ" ]
universe = ['SPY', 'V', 'PYPL', 'VZ', 'T', 'VIV', 'AMX', 'VOD', 'TU', 'TEF', 'TSLA',
           'GM', 'F', 'STLA', 'HMC', 'NIO', 'PG', 'KO', 'PEP', 'ABEV', 'NKE',
           'PM', 'UL', 'BUD', 'MDLZ', 'MO', 'DIS', 'CMCSA', 'SBUX', 'WBD',
           'BSX', 'HLN', 'AAPL', 'NVDA', 'TSM', 'CSCO', 'QCOM', 'RTX', 'AMD',
           'INTC', 'MU', 'NOK', 'XOM', 'PBR', 'CVX', 'SHEL', 'COP', 'TTE', 'BP', 
           'EQNR', 'CNQ', 'SU', 'UBS', 'JPM', 'BAC', 'ITUB', 'WFC', 
           'MS', 'HSBC', 'SCHW', 'BBD', 'JNJ', 'MRK', 'ABBV', 'PFE', 'NVO', 
           'NVS', 'AZN', 'ABT', 'BMY', 'SNY', 'ENB', 'SLB', 'EPD', 'KMI', 'ET', 
           'VALE', 'BHP', 'RIO', 'FCX', 'GOLD', 'SUZ', 'AMCR', 'GE', 'ABB', 
           'AMZN', 'WMT', 'HD', 'BABA', 'PDD', 'CVS', 'JD', 'TJX', 'CPNG', 
           'MSFT', 'GOOG', 'GOOGL', 'META', 'ORCL', 'BEKE', 'SAP', 'INFY', 
           'WIT', 'NU', 'CSX', 'UBER', 'GRAB', 'NEE', 'SO', 'PCG', 'MCD']

avg_sentiment_uni = []
portfolio_filter, avg_sentiment_port = [], []

for e in universe:
  _ticker_sent = []
  for headline in Quote(e).outer_news_df["Headline"]:
    sent = TextBlob(headline).sentiment.polarity
    if sent != 0:
      _ticker_sent.append(sent)
      print("{}\t{}\t{}".format(e, round(sent,2),headline))
  _avg_sent = np.average(_ticker_sent)
  print(">>> Avg ={}\n".format(_avg_sent))
  avg_sentiment_uni.append(_avg_sent)
  if e in portfolio:
    portfolio_filter.append(e)
    avg_sentiment_port.append(_avg_sent)
```

```python
col2_name = "Avg. Sentiment"
df = pd.DataFrame.from_dict({"Ticker":universe, col2_name:avg_sentiment_uni}) 
df.sort_values(col2_name, ascending=False).style.background_gradient()
```

```python
col2_name = "Avg. Sentiment"
df = pd.DataFrame.from_dict({"Ticker":portfolio_filter, col2_name:avg_sentiment_port}) 
df.sort_values(col2_name, ascending=False).style.background_gradient()
```
